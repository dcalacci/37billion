```python
def zip_fixer(num):
    if len(str(num))==4:
        return '0'+str(num)
    else:
        return str(num)
```

```python
rae['veh_zip_fix'] = rae['veh_zip'].apply(lambda x:zip_fixer(x))
rae.head()
```

```python
rae_gb_zip = rae.groupby('veh_zip')
```

```python
gb = rae_gb_zip.agg({'msrp':{'vehicles':len,
                             'msrp_mean':np.mean,
                             'msrp_min':min,
                             'msrp_max':max,
                             'msrp_std':np.std},
                     'mpg2008':{'mpg_mean':np.mean,
                                'mpg_min':min,
                                'mpg_max':max,
                                'mpg_std':np.std},
                     'curbwt':{'curbwt_mean':np.mean},
                     'hybrid':{'hybrid_frac':lambda x:sum(x)/float(len(x))*100.},
                     'mcycle':{'mcycle_frac':lambda x:sum(x)/float(len(x))*100.},
                     'veh_type':{'car_frac':lambda x:sum(x == 'Car')/float(len(x))*100,
                                 'truck_frac':lambda x:sum(x == 'Truck')/float(len(x))*100.,
                                 'suv_frac':lambda x:sum(x == 'SUV')/float(len(x))*100.},
                     'mi_per_day':{'mpd_mean':np.mean,
                                   'mpd_min':min,
                                   'mpd_max':max,
                                   'mpd_std':np.std}
                     })
gb.columns = gb.columns.droplevel(0)
gb = gb.reset_index()
gb_ma = gb[((int(mass_zipcodes[0][1]) >= gb['veh_zip']) & (int(mass_zipcodes[0][0]) <= gb['veh_zip'])) | ((int(mass_zipcodes[1][1]) >= gb['veh_zip']) & (int(mass_zipcodes[1][0]) <= gb['veh_zip']))]
gb_ma = gb.set_index('veh_zip')
gb_ma.head()
```

```python
def remove_crap(element):
    try:
        element = float(element)
    except ValueError:
        element = np.nan
    return element
```

```python
interesting_df = census_df[sorted(interesting.keys())]
interesting_df.columns = [interesting[i] for i in interesting_df.columns]
interesting_df = interesting_df.set_index('ZIP')
interesting_df = interesting_df.apply(lambda x: x.apply(remove_crap))
interesting_df.head()
```
